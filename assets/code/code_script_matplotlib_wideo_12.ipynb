#### Obraz w postaci liczbowej

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread('C:\\Users\\Andrey2\\Desktop\\python_image.png')

print(len(img))
print(len(img[0]))
print(len(img[0][0]))
print(img)
```

#### Wizualizacja tablic numpy jako obraz√≥w

```python
plt.figure(figsize=(10,15))
imgplot = plt.imshow(img)
```

#### Pseudokolor na wykresach z obrazami

```python
lum_img_channel_0 = img[:,:,0]
lum_img_channel_1 = img[:,:,1]
lum_img_channel_2 = img[:,:,2]

f, axs = plt.subplots(1,3, figsize=(15,15))

axs[0].imshow(lum_img_channel_0, cmap="summer")
axs[1].imshow(lum_img_channel_1, cmap="winter")
axs[2].imshow(lum_img_channel_2, cmap="hot")

plt.show()
```

```python
implot = plt.imshow(lum_img_channel_0, cmap="summer")
implot.set_cmap('nipy_spectral')
#implot.set_cmap('plasma')
#implot.set_cmap('magma')
```

```python
implot = plt.imshow(lum_img_channel_0, cmap="magma")
plt.colorbar(shrink=1.0)
```

#### Sprawdzenie specyficznych zakres√≥w obrazu

```python
plt.hist(lum_img_channel_0.ravel(), bins=64, range=(0.0,1.0), fc='k', ec='k')
plt.show()
```

```python
imgplot = plt.imshow(lum_img_channel_0,clim=(0.7,0.95))
```

```python
fig = plt.figure(figsize=(10,10))
a = fig.add_subplot(1,2,1)
imgplot = plt.imshow(lum_img_channel_0)
a.set_title('Do')
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal')
a = fig.add_subplot(1,2,2)
imgplot = plt.imshow(lum_img_channel_0)
imgplot.set_clim(0.0, 0.4)
a.set_title('Po')
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal')
```

#### Schematy interpolacji

```python
from PIL import Image

img = Image.open('C:\\Users\\Andrey2\\Desktop\\python_image.png')
img.thumbnail((64,64), Image.ANTIALIAS)

f, ax = plt.subplots(1,3,figsize=(15,15))
ax[0].imshow(img)
ax[0].set_title("bilinear")
ax[1].imshow(img, interpolation="nearest")
ax[1].set_title("nearest")
ax[2].imshow(img, interpolation="bicubic")
ax[2].set_title("bicubic")

plt.show()
```
