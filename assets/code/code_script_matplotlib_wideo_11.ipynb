#### Wykres z wype≈Çnieniem

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def get_coordinates(size):
    x = np.arange(-size,size, 0.001)
    x = np.append(x, size)
    y = [math.sqrt(size**2-value**2) for value in x]
    return x, y

plt.figure(figsize=(8,8))
plt.axis('equal')

x,y = get_coordinates(1.00)
plt.fill(x,y,c='r')

x,y = get_coordinates(0.75)
plt.fill(x,y,c='b')

x,y = get_coordinates(0.50)
plt.fill(x,y,c='w')

plt.show()
```

#### Wykres w uk≈Çadzie wsp√≥≈Çrzƒôdnych biegunowych

```python
import numpy as np
import matplotlib.pyplot as plt

xs = np.arange(0,20,0.01)
ys = xs**2

fig = plt.figure(figsize=(10,10))
plt.polar(xs,ys, 'b', linewidth=3)
plt.show()
```
