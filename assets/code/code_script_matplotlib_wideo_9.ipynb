### Wykres słupkowy

```python
from matplotlib.ticker import FuncFormatter
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(4)
money = [1.5e5, 2.5e6, 5.5e6, 2.0e7]

def millions(x, pos):
    'The two args are the value and tick position'
    return '$%1.1fM' % (x * 1e-6)

formatter = FuncFormatter(millions)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8,1))
plt.setp(ax, xticks=x, xticklabels=['Bill','Fred', 'Mary', 'Sue'])

ax[0].yaxis.set_major_formatter(formatter)
ax[0].bar(x,money)
ax[0].set_title("Oś Y z formatowaniem")

ax[1].bar(x,money)
ax[1].set_title("Osie Y i X bez formatowania")

plt.sca(ax[1])
plt.xticks(x,x)

plt.subplots_adjust(top=1.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25, wspace=0.35)
plt.show()
```

### Wykres kołowy

```python
import matplotlib.pyplot as plt

labels = 'Pióra', 'Długopisy', 'Ołówki', 'Markery'
sizes = [10,20,30,40]

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8,8))

explode = (0, 0.25, 0,0)
ax[0,0].pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
           shadow=True, startangle=0)
ax[0,0].axis('equal')

explode = (0, 0, 0.5,0)
ax[0,1].pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
           shadow=True, startangle=90)
ax[0,1].axis('equal')

explode = (0, 0.25, 0.25,0)
ax[1,0].pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
           shadow=True, startangle=180)
ax[1,0].axis('equal')

explode = (0.25, 0.25, 0.25,0.25)
ax[1,1].pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
           shadow=True, startangle=270)
ax[1,1].axis('equal')

plt.show()
```
