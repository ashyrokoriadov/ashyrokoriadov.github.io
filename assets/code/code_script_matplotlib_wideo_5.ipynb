### Praca z tekstem na wykresie

```python
import matplotlib.pyplot as plt
import numpy as np

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)

plt.xlabel('Dane o rozkładzie normalnym')
plt.ylabel('Prawdopodobieństwo')
plt.title('Histogram danych o rozkładzie normalnym')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40,160,0,0.03])
plt.grid(True)
plt.show()
```

### Praca z anotacjami

```python
ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t,s,lw=2)

plt.annotate('1 i 3 maksyma lokalne', xy=(0,1), xytext=(3, 1.5),
            arrowprops = dict(facecolor='black', shrink=0.05))

plt.annotate('1 i 3 maksyma lokalne', xy=(2,1), xytext=(3, 1.5),
            arrowprops = dict(facecolor='black', shrink=0.05))

plt.ylim(-2,2)
plt.show()
```

### Praca ze skalą logarytmiczną i innymi rodzajami skal

```python
from matplotlib.ticker import NullFormatter

np.random.seed(19680801)
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

plt.figure()

# skala liniowa
plt.subplot(221)
plt.plot(x,y)
plt.yscale('linear')
plt.title('skala liniowa')
plt.grid(True)

# skala logarytmiczna
plt.subplot(222)
plt.plot(x,y)
plt.yscale('log')
plt.title('skala logarytmiczna')
plt.grid(True)

# skala logarytmiczna symetryczna
plt.subplot(223)
plt.plot(x,y)
plt.yscale('symlog', linthreshy=0.01)
plt.title('skala logarytmiczna symetryczna')
plt.grid(True)

# skala logitowa
plt.subplot(224)
plt.plot(x,y)
plt.yscale('logit')
plt.title('skala logitowa')
plt.grid(True)
plt.gca().yaxis.set_minor_formatter(NullFormatter())

plt.subplots_adjust(top=1.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25, wspace=0.35)

plt.show()
```
