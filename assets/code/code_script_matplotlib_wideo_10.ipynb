#### Wykres z tabelą danych

```python
import numpy as np
import matplotlib.pyplot as plt

data = [[ 66386, 174296,  75131, 577908,  32015],
        [ 58230, 381139,  78045,  99308, 160454],
        [ 89135,  80552, 152558, 497981, 603535],
        [ 78415,  81858, 150656, 193263,  69638],
        [139361, 331509, 343164, 781380,  52269]]

columns = ('Mróz', 'Wiatr', 'Powódź', 'Trzęsienie ziemi', 'Gradobicie')
rows = ['%d lat' % x for x in (100,50,20,10,5)]

values = np.arange(0,2500,500)
value_increment = 1000

colors = plt.cm.YlGn(np.linspace(0,0.75,len(rows)))
n_rows=len(data)

index = np.arange(len(columns)) + 0.3
bar_width = 0.4

y_offset = np.zeros(len(columns))

cell_text = []
for row in range(n_rows):
    plt.bar(index, data[row], bottom=y_offset, color=colors[row])
    y_offset = y_offset + data[row]
    cell_text.append(['%1.1f' % (x / 1000.0) for x in y_offset])

colors = colors[::-1]
cell_text.reverse()

the_table = plt.table(cellText = cell_text,
                     rowLabels = rows,
                     rowColours = colors,
                     colLabels = columns,
                     loc='bottom')

plt.subplots_adjust(left=0.1, bottom=0.1)

plt.ylabel("Straty w ${0}".format(value_increment))
plt.yticks(values * value_increment, ['%d' % val for val in values])
plt.xticks([])
plt.title("Straty wyrządzone przez żywioły")

plt.show()
```

#### Wykres punktowy

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(19680801)

N=50
x = np.random.rand(N)
y = np.random.rand(N)
colors = plt.cm.coolwarm(np.linspace(0,1,N))
area = (30*np.random.rand(N))**2

plt.scatter(x,y, s=area, c=colors, alpha=0.75)
plt.show()
```
